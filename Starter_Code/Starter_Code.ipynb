```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#  YOUR CODE GOES HERE

nn = tf.keras.models.Sequential()

# First hidden layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units=64, activation='relu', input_dim=X_train_scaled.shape[1]))


# Second hidden layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units=32, activation='relu'))


# Output layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))


# Check the structure of the model (show reference)
nn.summary()

# I am not sure why this was not working in my last runs, it works now but the above cell is an absolute mess I combined too many steps
# Not sure best method to clean it up or which steps were missing in the other cells where X train was not being recognizd
```

```python
# Compile the model
#  YOUR CODE GOES HERE
#nn.compile(
    #optimzer="adam",
    #loss='Binary_crossentropy',
    #metrics=['accuracy'] # SyntaxError: incomplete input?

nn.compile(
    optimizer='adam',                
    loss='binary_crossentropy',     
    metrics=['accuracy']         
)

#running here
```

```python
# Train the model
#  YOUR CODE GOES HERE
history = nn.fit(
    X_train_scaled,
    y_train,
    epochs=10,
    batch_size=32,
    validation_data=(X_test_scaled, y_test),
    verbose=1
)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
